```python
import numpy as np
from pandas import DataFrame
import pandas as pd
milk_prod=pd.read_csv('milk_production.csv')
milk_prod
```

# Which state has produced max milk in year 2013-14?

```python
milk_prod['Total 10-11']=milk_prod['Cow Milk-2010-11']+milk_prod['Boffalo Milk-2010-11']+milk_prod['Goat Milk-2010-11']
milk_prod['Total 11-12']=milk_prod['Cow Milk-2011-12']+milk_prod['Boffalo Milk-2011-12']+milk_prod['Goat Milk-2011-12']
milk_prod['Total 13-14']=milk_prod['Cow Milk-2013-14']+milk_prod['Boffalo Milk-2013-14']+milk_prod['Goat Milk-2013-14']
milk_prod['Total 14-15']=milk_prod['Cow Milk-2014-15']+milk_prod['Boffalo Milk-2014-15']+milk_prod['Goat Milk-2014-15']
milk_prod['Total 15-16']=milk_prod['Cow Milk-2015-16']+milk_prod['Boffalo Milk-2015-16']+milk_prod['Goat Milk-2015-16']

milk_prod[milk_prod['Total 13-14']==milk_prod['Total 13-14'].max()][['State/ UT Name']]
```

# Calculate average milk production by cow, buffalo and goat in each year for selected state.

```python
st=input("Enter state ")
milk_prod['mean 10-11']=milk_prod[['Cow Milk-2010-11', 'Boffalo Milk-2010-11', 'Goat Milk-2010-11']].mean(1)
milk_prod['mean 11-12']=milk_prod[['Cow Milk-2011-12', 'Boffalo Milk-2011-12', 'Goat Milk-2011-12']].mean(1)
milk_prod['mean 13-14']=milk_prod[['Cow Milk-2013-14', 'Boffalo Milk-2013-14', 'Goat Milk-2013-14']].mean(1)
milk_prod['mean 14-15']=milk_prod[['Cow Milk-2014-15', 'Boffalo Milk-2014-15', 'Goat Milk-2014-15']].mean(1)
milk_prod['mean 15-16']=milk_prod[['Cow Milk-2015-16', 'Boffalo Milk-2015-16', 'Goat Milk-2015-16']].mean(1)
milk_prod[milk_prod['State/ UT Name']==st][['State/ UT Name','mean 10-11','mean 11-12','mean 13-14','mean 14-15','mean 15-16']]
```

# Draw line graph showing total milk production for last 5 years of selected state

```python
import matplotlib.pyplot as plt
st1=input("Enter state ")
df=milk_prod[milk_prod['State/ UT Name']==st1][['State/ UT Name','Total 10-11','Total 11-12','Total 13-14','Total 14-15','Total 15-16']]

#x=[df['Total 10-11'],df['Total 11-12'],df['Total 13-14'],df['Total 14-15'],df['Total 10-11']]
plt.plot(df['Total 10-11'],df['State/ UT Name'], linestyle="dotted", marker='^', color="red", label = 'Total 10-11')
plt.plot(df['Total 11-12'],df['State/ UT Name'], linestyle="dotted", marker='o', color="red", label = 'Total 11-12')
plt.plot(df['Total 13-14'],df['State/ UT Name'], linestyle="dotted", marker='<', color="red", label = 'Total 13-14')
plt.plot(df['Total 14-15'],df['State/ UT Name'], linestyle="dotted", marker='>', color="red", label = 'Total 14-15')
plt.plot(df['Total 15-16'],df['State/ UT Name'], linestyle="dotted", marker='*', color="red", label = 'Total 15-16')
plt.legend()
plt.show()

```
