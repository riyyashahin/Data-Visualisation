# List the countries having maximum oil reserves in each region.

```python
df=oil_r.sort_values(['2014','2015','2016'] ,ascending=False)
df
df_NA=df[df['Region']=='North America']
df_SC=df[df['Region']=='S. & Cent. America']
df_EU=df[df['Region']=='Europe & Eurasia'] 
df_ME=df[df['Region']=='Middle East']
df_A=df[df['Region']=='Africa']
df_AS=df[df['Region']=='Asia Pacific']
print(df_NA.head(1)[['Region','Country/ Region']])
print(df_SC.head(1)[['Region','Country/ Region']])
print(df_EU.head(1)[['Region','Country/ Region']])
print(df_ME.head(1)[['Region','Country/ Region']])
print(df_A.head(1)[['Region','Country/ Region']])
print(df_AS.head(1)[['Region','Country/ Region']])
```
